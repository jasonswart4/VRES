First I want to solve the simple harmonic oscillator using a numerical method. 

$\ddot{x}+\frac{b}{m}\dot{x}+\frac{k}{m}x=0$

Finite difference approximation is given by 

$\dot{x}\approx\frac{x(t+\Delta t)-x(t)}{\Delta t}$

$\ddot{x}\approx \frac{x(t+\Delta t) - 2x(t) + x(t-\Delta t)}{\Delta t^2}$

```python
k = 10
m = 2
b = 1

n_points = 1000 # number of points
t0 = 0 # initial time 
tf = 10 # final time 
dt = (tf - t0) / n_points # Delta t

# Initial conditions
x0 = 0
v0 = 1

x = n_points*[0]

x[0] = x0
x[1] = x0 + v0*dt

for i in range(n_points-2):
    x[i+2] = (-k/m * x[i+1] + b*x[i+1] / (m*dt) + (2*x[i+1] - x[i])/dt**2) / (1/dt**2 + b/(m*dt))

plt.plot(x)
plt.show()
```
